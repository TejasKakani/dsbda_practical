```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#1
df = pd.read_csv('Downloads/demo1.csv')
#2
df.boxplot(column=['math score', 'reading score'])
plt.show()

#3
q10 = df.quantile(0.10)
q90 = df.quantile(0.90)

df['math score'] = np.where(df['math score'] < q10['math score'], q10['math score'], df['math score'])
df['math score'] = np.where(df['math score'] > q90['math score'], q90['math score'], df['math score'])

df.boxplot(column=['math score', 'reading score'])
plt.show()
```
