```python
import pandas as pd
import numpy as np

def detect_outliers_iqr(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return dataframe[(dataframe[column] < lower_bound) | (dataframe[column] > upper_bound)]

def remove_outliers_iqr(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return dataframe[(dataframe[column] >= lower_bound) & (dataframe[column] <= upper_bound)]

#1
df = pd.read_csv('Downloads/demo1.csv')
df

#2
outliers = detect_outliers_iqr(df, 'writing score')
print("Outliers:\n", outliers)

#3
df_cleaned = remove_outliers_iqr(df, 'writing score')
print("Cleaned:\n", df_cleaned)

print("\nOriginal dataset size:", df.shape)
print("Cleaned dataset size:", df_cleaned.shape)
```
