```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def floor_cap_outliers(df, column, lower_quantile=0.10, upper_quantile=0.90, factor=1.5):
    lower_bound = df[column].quantile(lower_quantile) - factor
    upper_bound = df[column].quantile(upper_quantile) + factor
    df[column] = np.where(df[column] < lower_bound, lower_bound,
                          np.where(df[column] > upper_bound, upper_bound, df[column]))
    return df

#1
df = pd.read_csv('Downloads/demo1.csv')
df

#2
plt.scatter(df['writing score'], df['placement score'], alpha=0.4)
plt.show()

#3
df_handled = floor_cap_outliers(df.copy(), 'writing score')
plt.scatter(df_handled['writing score'], df_handled['placement score'], alpha=0.4)
plt.show()

plt.hist(df_handled['writing score'], bins=20, color='orange', edgecolor='black')
plt.title('After Flooring and Capping')
```
